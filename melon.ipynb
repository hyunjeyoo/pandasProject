```python
url1 = 'http://www.melon.com/chart/index.htm'
driver.get(url1)
```

```python
from bs4 import BeautifulSoup
# 전체 html 코드를 갖고 오는 기능 실행
html1 = driver.page_source
# BeautifulSoup(html1, 'html.parser')를 soup1으로 줄이기
soup1 = BeautifulSoup(html1, 'html.parser')
```

```python
soup1
```

```python
songs = soup1.select('tr')[1:]
print(len(songs))
print(songs[0])
```

```python
song = songs[0]
```

```python
# <a> </a> 사이에 있는 html 코드를 가져오기
title = song.select('a')
len(title)
```

```python
title = song.select('div.ellipsis.rank01 > span > a')
# div ~~~ ellipsis rank01 ~~~ 인데, '.'으로 표기한다. 띄어쓰기는 '_' 표시
title
```

```python
# html 코드 중 텍스트를 가져오기
title = song.select('div.ellipsis.rank01 > span > a')[0].text
title
```

```python
singer = song.select('div.ellipsis.rank02 > a')[0].text
singer
```

```python
# 노래 100개의 제목과 가수 출력하기
# 'for x in range()' 형식: songs가 100개인데 0번째 song 출력하고, 1번째 song 출력하고, ... , 99번째 song 출력.
for song in songs:
    title = title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    print(title, singer, sep = '|')
```

```python
song_data = []
rank = 1

songs = soup1.select('tr')[1:]
for song in songs:
    title = title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    song_data.append(['Melon', rank, title, singer]) #.append는 '이어붙이다'
    rank = rank + 1

song_data
```

```python
# 데이터분석 라이브러리 pandas 가져오기
import pandas as pd
# 엑셀 파일에서 열 종류 이름 정리
column = ['서비스', '순위', '타이틀', '가수']
# 데이터 프레임 생성하기
pd_data = pd.DataFrame(song_data, columns = column)
# 생성된 데이터 프레임 상위 5개 확인하기
pd_data.head()
```
