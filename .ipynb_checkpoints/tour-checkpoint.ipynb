```python
#데이터 전처리 함수
#def create_kto_date든 뭐든 이름은 마음대로
def create_kto_data(yy,mm):
    # 1. 불러올 엑셀 파일 경로 지정
    file_path = './files/tour_data/kto_{}{}.xlsx'.format(yy,mm)
    
    # 2. 엑셀 파일 불러오기
    df = pd.read_excel(file_path, header = 1, skipfooter = 4, usecols = 'A:G')
    
    # 3. 기준 년월 칼럼 추가
    df['기준 년월'] = '{}-{}'.format(yy,mm)
    
    # 4. 국적 칼럼에서 대륙 제거하고 국가만 남기기
    ignore_list = ['아시아주', '미주', '구주', '아프리카주', '대양주', '기타대륙', '교포소계']
    condition = (df['국적'].isin(ignore_list) == False) # 대륙 미포함 조건
    df_country = df[condition].reset_index(drop=True)
    
    # 5. '대륙'이란 칼럼을 추가하며 대륙 정보 넣기
    continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3 + ['아프리카']*2 + ['기타 대륙'] + ['교포']
    df_country['대륙'] = continents
    
    # 6. 국가별 '관광객 비율(%)'이란 칼럼 추가하며 수치 넣기
    df_country['관광객 비율(%)'] = round(df_country.관광 / df_country.계 * 100, 1)
        #df_country.관광은 한 개씩 연산할 때 씀
        #df_country['관광']은 칼럼 전체를 가져오거나 새로 만들 때 씀
    
    # 7. '전체 비율(%)'이란 칼럼 추가하며 수치 넣기
    tourist_sum = sum(df_country['관광'])
    df_country['전체 비율(%)'] = round(df_country['관광'] / tourist_sum * 100, 1)
    
    # 8. 결과 출력
    return(df_country)
```

```python
#이 모든 파일의 데이터를 한 파일로 통합
df = pd.DataFrame()
```

```python
for yy in range(2010, 2021):
    for mm in range(1, 13):
        #한 달어치의 데이터를 가진 파일을 열어 전처리 함수를 한 번 돌릴 때마다 'temp' 변수에 저장
        temp = create_kto_data(str(yy), str(mm).zfill(2))
        #df에 temp를 붙여서 df로 저장하고, 또 저장된 df에 temp를 붙여서 df로 다시 저장하고, 또다시 df에 temp를...
        df = pd.concat([df, temp], ignore_index=True)
        #컴퓨터는 단순하기 때문에 202006이 없다고 찾고 찾으며 노란색 표시를 반복적으로 나타내다가 맨 끝에 202006이 없다고 표시할 것임
```
